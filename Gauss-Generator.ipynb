```python
import numpy as np
import matplotlib.pyplot as plt

def generate_arterial_pulse(length=300, p1_center=100, p1_width=20, p2_center=180, p2_width=30):
    x = np.arange(length)
    g1 = np.exp(-0.5 * ((x - p1_center) / p1_width) ** 2)
    g2 = 0.5 * np.exp(-0.5 * ((x - p2_center) / p2_width) ** 2)
    pulse = g1 + g2
    return pulse

def generate_synthetic_signal(num_pulses=60, length_per_pulse=300):
    signal = np.concatenate([generate_arterial_pulse(length_per_pulse) for _ in range(num_pulses)])
    return signal

signal = generate_synthetic_signal(num_pulses=10, length_per_pulse=300)
plt.figure(figsize=(15, 3))
plt.plot(signal, linewidth=1)
plt.title("Synthetic Arterial Signal (Gaussian Pulse-Based)")
plt.xlabel("Time (samples)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

P1_CENTER_RANGE = (90, 110)
P1_WIDTH_RANGE  = (15, 25)
P2_CENTER_RANGE = (170, 190)
P2_WIDTH_RANGE  = (25, 35)
A1_RANGE        = (0.8, 1.2)
A2_RANGE        = (0.3, 0.7)

def generate_arterial_pulse(length=300):
    x = np.arange(length)
    p1c = np.random.uniform(*P1_CENTER_RANGE)
    p1w = np.random.uniform(*P1_WIDTH_RANGE)
    p2c = np.random.uniform(*P2_CENTER_RANGE)
    p2w = np.random.uniform(*P2_WIDTH_RANGE)
    a1  = np.random.uniform(*A1_RANGE)
    a2  = np.random.uniform(*A2_RANGE)
    g1  = a1 * np.exp(-0.5 * ((x - p1c) / p1w) ** 2)
    g2  = a2 * np.exp(-0.5 * ((x - p2c) / p2w) ** 2)
    return g1 + g2

def generate_synthetic_signal(num_pulses=60, length_per_pulse=300):
    return np.concatenate([generate_arterial_pulse(length_per_pulse) for _ in range(num_pulses)])

signal = generate_synthetic_signal(num_pulses=10, length_per_pulse=300)
plt.figure(figsize=(15, 3))
plt.plot(signal, linewidth=1)
plt.title("Synthetic Arterial Signal with Parameter Variation")
plt.xlabel("Time (samples)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.tight_layout()
plt.show()
```
